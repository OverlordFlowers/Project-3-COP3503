```python
df = pd.read_csv("songs.csv", index_col='name', parse_dates=['release_date'], encoding = "ISO-8859-1")
```

```python
df.head(5)
```

```python
danceabilityModifier = 10000.0 * 1.2
energyModifier = 10000.0 * .5
valenceModifier = 10000.0 * 1.3
tempoModifier = 0
new_col = round((((df['danceability']*danceabilityModifier)+(df['energy']*energyModifier)+(df['valence']*valenceModifier))/3)+(df['tempo']*tempoModifier)).tolist()
```

```python
df.insert(loc=18, column='emotionVal', value=new_col)
df.to_csv('songsEmotive.csv')
```
